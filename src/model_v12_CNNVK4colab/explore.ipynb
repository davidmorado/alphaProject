```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
#from CNN_VK import CNN_VK
#from varkeys import Varkeys
```

```python
l = []
for filename in os.listdir('gridresults'):
    with open('gridresults/'+filename, 'rb') as f:
        l.append(pickle.load(f))
```

```python
l[0][0]
```

```python
MaxAccPerTP = {tp:0 for tp in [0.1, 0.2, 0.3, 0.5, 0.75, 1.]}
BestHpPerTP = {tp:() for tp in [0.1, 0.2, 0.3, 0.5, 0.75, 1.]}
BestVAPerTP = {tp:[] for tp in [0.1, 0.2, 0.3, 0.5, 0.75, 1.]}
BestHPacc = {tp:0 for tp in [0.1, 0.2, 0.3, 0.5, 0.75, 1.]}
BBW = None
BKPC = None
```

```python
for model in l:
    ctp = model[0]['train_percentage']
    chp = model[0]
    cbva = model[1]['val_acc']
    cmaxva = max(model[1]['val_acc'])
    if cmaxva > MaxAccPerTP[ctp]:
        BestHpPerTP[ctp] = chp
        BestVAPerTP[ctp] = cbva
        MaxAccPerTP[ctp] = cmaxva

BBW = BestHpPerTP[1.]['bandwidth']
BKPC = BestHpPerTP[1.]['n_keys_per_class']
for model in l:
    ctp = model[0]['train_percentage']
    cbw = model[0]['bandwidth']
    ckpc = model[0]['n_keys_per_class']
    cmaxva = max(model[1]['val_acc'])
    if cbw == BBW and ckpc == BKPC:
        BestHPacc[ctp] = cmaxva
```
