```python
import pickle 
import matplotlib.pyplot as plt
```

```python
d = pickle.load(open('best_metrics.pickle', 'rb'))
```

```python
CNN_acc_train_val = d['train_acc']
CNN_acc_test = d['test_acc']
SC_acc_train_val = d['comb_acc_train_val']
SC_acc_test = d['comb_acc_test']
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
#ax.plot(CNN_acc_train_val, label='CNN|train')
ax.plot(CNN_acc_test, label='CNN|test')
#ax.plot(SC_acc_train_val, label='SC|train')
ax.plot(SC_acc_test, label='SC|test')
ax.set_xticks(range(0, len(SC_acc_test)))
ax.set_xticklabels(range(1, len(SC_acc_test)+1))
ax.grid()
ax.set_ylim(0.1, 1.0)
ax.set_title('SimpleCache Learning Curve')
ax.set_ylabel('Accuracy')
ax.set_xlabel('Epochs')
ax.legend()
fig.show()
print('Final CNN accuracy:')
print(d['test_acc'][-1])
print('Final Memory accuracy:')
print(d['mem_acc_test'][-1])
print('Final SC accuracy:')
print(d['comb_acc_test'][-1])
```

```python
# find optimal hyperparameters
metrics_dict = pickle.load(open('metrics_dict.pickle', 'rb'))
best_hp = ()
best_acc = 0
for key in metrics_dict.keys():
    new_acc=metrics_dict[key]['comb_acc_val'][0]
    if new_acc > best_acc:
        best_acc = new_acc
        best_hp = key
print('best_hp')
print(best_hp)
print('best_acc')
print(best_acc)
```
